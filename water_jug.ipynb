```python
# Water Jug Problem with 8 User Rules and Step Count

# Get jug capacities and goal
jug1_capacity = int(input("Enter capacity of Jug A: "))
jug2_capacity = int(input("Enter capacity of Jug B: "))
goal = int(input("Enter goal amount: "))

# Initialize jugs
jug1 = 0
jug2 = 0
steps = 0  # Step counter

print("\n--- Rules ---")
print("1: Fill Jug A")
print("2: Fill Jug B")
print("3: Empty Jug A")
print("4: Empty Jug B")
print("5: Pour Jug A -> Jug B (all)")
print("6: Pour Jug B -> Jug A (all)")
print("7: Pour Jug A -> Jug B (partial)")
print("8: Pour Jug B -> Jug A (partial)")

print("\n--- Start ---")

# Loop until goal is reached
while jug1 != goal and jug2 != goal:
    print(f"\nCurrent state -> Jug A: {jug1}, Jug B: {jug2}")
    rule = input("Choose a rule (1-8): ")

    if rule not in [str(i) for i in range(1, 9)]:
        print("Invalid choice! Try again.")
        continue  # Don't count invalid step

    steps += 1  # Count valid steps

    if rule == "1":
        jug1 = jug1_capacity
        print(f"Step {steps}: Fill Jug A -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "2":
        jug2 = jug2_capacity
        print(f"Step {steps}: Fill Jug B -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "3":
        jug1 = 0
        print(f"Step {steps}: Empty Jug A -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "4":
        jug2 = 0
        print(f"Step {steps}: Empty Jug B -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "5":  # Pour all Jug A -> Jug B
        jug2 += jug1
        jug1 = 0
        if jug2 > jug2_capacity:
            jug1 = jug2 - jug2_capacity
            jug2 = jug2_capacity
        print(f"Step {steps}: Pour Jug A -> Jug B (all) -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "6":  # Pour all Jug B -> Jug A
        jug1 += jug2
        jug2 = 0
        if jug1 > jug1_capacity:
            jug2 = jug1 - jug1_capacity
            jug1 = jug1_capacity
        print(f"Step {steps}: Pour Jug B -> Jug A (all) -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "7":  # Partial Pour Jug A -> Jug B
        transfer = min(jug1, jug2_capacity - jug2)
        jug1 -= transfer
        jug2 += transfer
        print(f"Step {steps}: Partial Pour Jug A -> Jug B ({transfer}) -> Jug A: {jug1}, Jug B: {jug2}")
    elif rule == "8":  # Partial Pour Jug B -> Jug A
        transfer = min(jug2, jug1_capacity - jug1)
        jug2 -= transfer
        jug1 += transfer
        print(f"Step {steps}: Partial Pour Jug B -> Jug A ({transfer}) -> Jug A: {jug1}, Jug B: {jug2}")

print(f"\n Goal reached in {steps} steps! Final state -> Jug A: {jug1}, Jug B: {jug2}")
```
